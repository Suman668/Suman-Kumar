# Which month have highest sales

```python
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int32')
data.head()
```

# Add a sales Column

```python
# for i in data.columns:
#     data[i]=pd.to_numeric(data[i],errors = 'ignore')
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])
```

```python
data['Sales']= data['Quantity Ordered']*data['Price Each']
data
```

```python
Monthly_sales = data.groupby('Month').sum()
Monthly_sales
```

# Which City has highest Sales

```python
data.head(4)
```

```python
# Add a city column

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

data['City'] = data['Purchase Address'].apply(lambda x : get_city(x)+' ('+get_state(x)+')')
data.head(4)
```

```python
City_wise_sales = data.groupby('City').sum()
City_wise_sales
```

```python
plt.figure(figsize=(16,8))

cities = [city for city , df in data.groupby('City')]

plt.bar(cities,City_wise_sales['Sales'])
plt.xticks(cities,rotation = 'vertical',size = 12)
plt.yticks(City_wise_sales['Sales'], size = 12)
plt.xlabel('City Names', size = 16)
plt.ylabel('Sales in $',size = 16)
```
